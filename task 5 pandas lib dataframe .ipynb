```python
# creat dataframe using matrix
import pandas as pd

# Define your matrix
matrix = [
    ['John', 25, 'Male'],
    ['Alice', 30, 'Female'],
    ['Bob', 35, 'Male']
]

# Define column names
columns = ['Name', 'Age', 'Gender']

# Create DataFrame
df_matrix_1 = pd.DataFrame(matrix, columns=columns)

# Display DataFrame
print("DataFrame 1 created using a matrix:")
print(df_matrix_1)
```

```python
# creat dataframe with dictionary
import pandas as pd

# Define your dictionary
data = {
    'Name': ['John', 'Alice', 'Bob'],
    'Age': [25, 30, 35],
    'Gender': ['Male', 'Female', 'Male']
}

# Create DataFrame
df_dict_1 = pd.DataFrame(data)

# Display DataFrame
print("\nDataFrame 1 created using a dictionary:")
print(df_dict_1)
```
